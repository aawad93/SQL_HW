```python
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine


engine = create_engine('postgresql://postgres:andrew2005@localhost/HomeworkDB')
connection = engine.connect()
salaries_df = pd.read_sql('Salaries', connection)
titles_df = pd.read_sql('Titles', connection)
employees_df = pd.read_sql('Employees', connection)
```

```python
df = pd.merge(salaries_df, titles_df, on = 'emp_no')

```

```python
grouped_df = df.groupby(by='title')
df_avg = grouped_df['salary'].mean()

df_avg.plot(kind='bar')
```
